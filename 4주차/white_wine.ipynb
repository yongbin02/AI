```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/AI/data/winequality-white.csv", delimiter=';').dropna()
df.head()
```

```python
# 데이터 라벨링
# 7점 이상 -> 좋은 와인(1) / 6점 이하 -> 보통 와인(0)
df['quality'] = df['quality'].apply(lambda x: 1 if x > 6 else 0)
print("라벨링 방식: 이진 분류 (0: 보통 와인, 1: 좋은 와인)")

# 데이터 라벨링 이후 분포 확인
print(df['quality'].value_counts())
print("\n")
```

```python
# 라벨링 이후 데이터 확인
df.head()
```

```python
# 독립변수 및 종속변수 분리
X = df.drop(columns=["quality"])
y = df["quality"]

# 학습 데이터 및 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# 데이터 스케일링
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# ------------------
# 1) KNN 구하기
# ------------------

# 1. KNN 모델 생성 (이웃의 수 5로 설정)
knn_model = KNeighborsClassifier(n_neighbors=5)

# 2. 모델 학습 시키기
knn_model.fit(X_train, y_train)

# 3. 성능 평가하기
y_pred = knn_model.predict(X_test)

# 4. 정답과 모델의 예측값 비교
accuracy_knn = accuracy_score(y_test, y_pred)

print(f"KNN 모델의 예측 정확도: {accuracy_knn:.4f}")
```

```python
# ----------------------------
# 2) 랜덤포레스트 구하기
# ----------------------------

# 1. 랜덤 포레스트 모델 생성
rf_model = RandomForestClassifier(random_state=42)

# 2. 모델 학습 시키기
rf_model.fit(X_train, y_train)

# 3. 성능 평가하기
y_pred = rf_model.predict(X_test)

# 4. 정답과 모델의 예측값 비교
accuracy_rf = accuracy_score(y_test, y_pred)

print(f"랜덤 포레스트 모델의 예측 정확도: {accuracy_rf:.4f}")
```

```python
# -----------------------------
# 3) 결정 트리 구하기
# -----------------------------

# 1. 결정 트리 모델 생성
dt_model = DecisionTreeClassifier(random_state=42)

# 2. 모델 학습 시키기
dt_model.fit(X_train, y_train)

# 3. 성능 평가하기
y_pred = dt_model.predict(X_test)

# 4. 정답과 모델의 예측값 비교
accuracy_dt = accuracy_score(y_test, y_pred)

print(f"결정 트리 모델의 예측 정확도: {accuracy_dt:.4f}")
```

```python
# -----------------------------
# 4) 로지스틱 회귀 구하기
# -----------------------------

# 1. 로지스틱 회귀 모델 생성
lr_model = LogisticRegression()

# 2. 모델 학습 시키기
lr_model.fit(X_train, y_train)

# 3. 성능 평가하기
y_pred = lr_model.predict(X_test)

# 4. 정답과 모델의 예측값 비교
accuracy_rg = accuracy_score(y_test, y_pred)

print(f"로지스틱 회귀 모델의 예측 정확도: {accuracy_rg:.4f}")
```

```python
# 전체 모델 정확도 비교
print(f"KNN 모델의 예측 정확도: {accuracy_knn:.4f}")
print(f"랜덤 포레스트 모델의 예측 정확도: {accuracy_rf:.4f}")
print(f"결정 트리 모델의 예측 정확도: {accuracy_dt:.4f}")
print(f"로지스틱 회귀 모델의 예측 정확도: {accuracy_rg:.4f}")
```

```python
# 로지스틱 회귀 상세 값
print("\n--- 로지스틱 회귀 모델 상세 값 ---")

# 절편 값 확인
intercept = lr_model.intercept_
print(f"절편 (Intercept): {intercept[0]:.4f}")

# 각 특성별 기울기(계수) 값 확인
coefficients = lr_model.coef_[0]
feature_names = X.columns

print("\n각 특성별 기울기 (Coefficients):")
for feature, coef in zip(feature_names, coefficients):
    print(f"- {feature}: {coef:.4f}")
```
